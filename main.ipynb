```python
# Data Cleaning and Preprocessing EDA of useful features only
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['user/gender'].fillna(df['user/gender'].mode()[0], inplace=True)
df['user/profileName'].fillna(df['user/profileName'].mode()[0], inplace=True)

df['review/text'].fillna('No Review', inplace=True)
df['review/timeUnix'] = pd.to_datetime(df['review/timeUnix'])
```

```python
df.info()
```

```python
# Text Preprocessing Cleaning with NLP/NLTK
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('all')
```

```python
def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    text = text.lower()
    text = ' '.join(WordNetLemmatizer().lemmatize(word) for word in text.split() if word not in set(stopwords.words('english')))
    return text
```

```python
df['review/text'] = df['review/text'].apply(clean_text)
df[['review/text']].head()
```

```python
# Feature Engineering and using TF-IDF Vectorization to Change Text Data into Numerical Representation
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=300)
text_feat = tfidf.fit_transform(df['review/text']).toarray()

numerical_feat = df[['beer/ABV', 'review/appearance', 'review/aroma', 'review/palate', 'review/taste']]
x = np.hstack((text_feat, numerical_feat))

y = df['review/overall']
```

```python
# we will use Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

lr = LinearRegression()
lr.fit(x_train, y_train)

# Prediction
y_pred = lr.predict(x_test)
print("Linear Regression RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("Linear Regression MAE:", mean_absolute_error(y_test, y_pred))
print("Linear Regression R²:", r2_score(y_test, y_pred))
```

```python
# Now we will use Random Forest Regression
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state=42)
rf.fit(x_train, y_train)

y_prediction = rf.predict(x_test)
print("Random Forest RMSE:", np.sqrt(mean_squared_error(y_test, y_prediction)))
print("Random Forest MAE:", mean_absolute_error(y_test, y_prediction))
print("Random Forest R²:", r2_score(y_test, y_prediction))
```
